```python
df['Category'] = "c"
```

```python
#AGRESSION SEXUELLES
sex_crime=df['Most Serious Crime'].str.contains("SEXUAL|RAPE|SEX|FEMALE|GENITAL|VIOL|INCEST|PORNOGRAPHY|FORCIBLE TOUCHING|INCEST|PORN|LURING CHLD TO B FEL|LURING CHLD TO A FEL", na=False)
df.loc[sex_crime, 'Category'] = "Sex Crime"
```

```python
#agression sur les jeunes
YOUTHFUL_offender= df['Most Serious Crime'].str.contains("YO-JO|YOUTHFUL", na=False)
df.loc[YOUTHFUL_offender, 'Category'] = "Youthful Offender"
```

```python
#VOLS
robery_crime=df['Most Serious Crime'].str.contains("BURGLARY|ROBBERY|LARCENY|LAR|STOL|CARD", na=False )
df.loc[robery_crime, 'Category'] = "Roberry"
```

```python
#FRAUDES
fraud= df['Most Serious Crime'].str.contains("FALSLY|FALSE|FRAUDULENT|FRAUDULNT|FRAUD|DEFRAUD|USURY|TAMPERING|TAMP|TRESSPAS|TAX|COERCION|BRIBING|BUS|BRIBERY|BRIBE|BAIL|STRIPPING|STALKING|TX|MONEY|IDENTITY|FORGERY|FORGE|ILLEGAL|ALCOHOL|TRADEMARK|TRADE|UNLAW POSS PERS ID 1|CRIM IMPERSONATION|UNLAW POSS PERS ID 2|LABOR TRAFFICKING|UNSTMPD CIGS|POSS SKIMMER DEVICE|MISUSE|SELL/ISS. INST. W/O LIC|REP FAIL FILE CORP", na=False)
df.loc[fraud, 'Category'] = "Fraud"
```

```python
#ASSOCIATION MALFAITEURS
conspiracy= df['Most Serious Crime'].str.contains("CONSPIRACY|SOLI|FACIL", na=False)
df.loc[conspiracy, 'Category'] = "Conspiracy"
```

```python
#TERRORISME
terrorism= df['Most Serious Crime'].str.contains("BOMB|TERROR", na=False)
df.loc[terrorism, 'Category'] = "Terrorism"
```

```python
#HOMICIDE
homicide=df['Most Serious Crime'].str.contains("MANSLAUGHTER|MURDER|STRANGULATION|HOMICD|CRIM NEGLI HOMICIDE|YO NEG HOMICIDE|CONCEALMENT HUMAN CORPSE|DIVERS PRESCRIP|DIVERS.PRESCRIP|FELONY", na=False)
df.loc[homicide, 'Category'] = "Homicide"
```
